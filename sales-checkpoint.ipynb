```python
import pandas as pd
from sqlalchemy import create_engine

# Load original file
df = pd.read_csv(r"C:\Users\Yashwanth\OneDrive\Documents\superstore\Sample - Superstore.csv", encoding='latin-1')
print("Original rows:", len(df))

# Drop empty rows BEFORE renaming columns
df = df.dropna(subset=['Order ID'])
print("After dropping empty rows:", len(df))

# NOW clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')

# Fix date columns
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

# Add date columns
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.strftime('%b')

# Fix postal code
df['postal_code'] = df['postal_code'].astype(str)

# Remove duplicates
df = df.drop_duplicates()

# Confirm
print("Final clean rows:", len(df))
print("Null values:", df.isnull().sum().sum())

# Export clean CSV
df.to_csv("clean_superstore.csv", index=False)
print("â clean_superstore.csv saved!")

# Export to Excel
df.to_excel("superstore_analysis.xlsx", index=False, sheet_name="Data")
print("â superstore_analysis.xlsx saved!")

# Reload PostgreSQL
engine = create_engine("postgresql://postgres:9535539228@localhost:5432/superstore")
df.to_sql("orders", engine, if_exists="replace", index=False)
print("â PostgreSQL reloaded!")
```
